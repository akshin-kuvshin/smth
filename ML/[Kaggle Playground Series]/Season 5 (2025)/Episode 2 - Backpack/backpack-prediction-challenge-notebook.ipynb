```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('/kaggle/input/playground-series-s5e2/train.csv')
X = X.drop(columns='id')
X
```

```python
X.isna().sum()
```

```python
X['Size'] = X['Size'].fillna('Medium')
X['Laptop Compartment'] = X['Laptop Compartment'].fillna('No')
X['Waterproof'] = X['Waterproof'].fillna('No')
X.isna().sum()
```

```python
X = X.dropna().reset_index(drop=True)
```

```python
X['Size'].value_counts()
```

```python
size_to_int = {
    'Small': 1,
    'Medium': 2,
    'Large': 3
}
X['Size'] = X['Size'].map(size_to_int)
```

```python
X
```

```python
X['Compartments'].value_counts()
```

```python
X['Compartments'] = X['Compartments'].astype(int)
```

```python
bool_to_int = {
    'No': 0,
    'Yes': 1
}
X['Laptop Compartment'] = X['Laptop Compartment'].map(bool_to_int)
X['Waterproof'] = X['Waterproof'].map(bool_to_int)
X
```

```python
from catboost import CatBoostRegressor
```

```python
y = X['Price']
X = X.drop(columns='Price')
```

```python
model = CatBoostRegressor(
    iterations=666,
    learning_rate = 0.0666,
    depth=10
)
model.fit(
    X,
    y,
    cat_features=['Brand', 'Material', 'Style', 'Color'],
    verbose=18
)
```

```python
pred_y = model.predict(X)
```

```python
from math import sqrt
def RMSE(la, lb):
    error = 0
    for a, b in zip(la, lb):
        error += abs(a - b)
    error = sqrt(error / len(la))
    return error
RMSE(y, pred_y)
```
