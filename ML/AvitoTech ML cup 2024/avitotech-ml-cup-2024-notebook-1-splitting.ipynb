```python
import pandas as pd
print('Setup completed!')
```

```python
train = pd.read_parquet('/kaggle/input/avitotech-ml-cup-2024-train-dataset')
campaigns = pd.read_parquet('/kaggle/input/avitotech-ml-cup-2024-campaigns-meta-dataset')
print('\'Train\' and \'Campaigns\' datasets have been successfully read!')
```

```python
train['target'] = train['target'].astype(bool)
train['event_date'] = train['event_date'].astype('datetime64[ns]')
print('success')
```

```python
campaigns['start_date'] = campaigns['start_date'].astype('datetime64[ns]')
campaigns['end_date'] = campaigns['end_date'].astype('datetime64[ns]')
print('success')
```

```python
train = train.merge(campaigns, on='adv_campaign_id', how='inner')
train.head(10)
```

```python
del campaigns
print('\'Campaigns\' dataset has been deleted.')
```

```python
train['start_date'] = ((train['event_date'] - train['start_date']) / 86400000000000).astype(int)
train['end_date'] = ((train['end_date'] - train['event_date']) / 86400000000000).astype(int)
train['goal_cost'] = train['goal_budget'] / train['goal_cost']
train.head(10)
```

```python
i = 0
step = 10_000_000
ind = 1
while i < len(train):
    j = i + step
    name = f'train_{ind}'
    train_part = train.iloc[i:j]
    train_part = train_part.drop(columns=['event_date']) \
                           .rename(columns={
                               'start_date': 'days_from_start',
                               'end_date'  : 'days_to_end',
                               'goal_cost' : 'goals_number'
                           })
    train_part.to_parquet(f'/kaggle/working/{name}.parquet')
    del train_part
    print(f'\'{name}\' dataset has been formed and written.')
    i += step
    ind += 1
```
